```python
inverses = {letter: 1 / values[0] for letter, values in letters.items()}
inverses_total = sum(inverses.values())
letters_probs = {letter: inv / inverses_total for letter, inv in inverses.items()}
letters_probs
```

```python
# isolate letters and probabilities
let = list(letters_probs.keys())
prob = list(letters_probs.values())
```

```python
# making the function
def letter_gen(layer = None):
	#select letters based on probabilities
	outer = np.random.choice(let, size = 16, p = prob).tolist()
	middle = np.random.choice(let, size = 8, p = prob).tolist()
	inner = np.random.choice(let, size = 1, p = prob).tolist()

	if layer == "outer":
		return outer

	elif layer == "middle":
		return middle

	elif layer == "inner":
		return inner
        
	elif layer == "all":
		return outer, middle, inner
        
	else:
		print('Layer must be specified as "outer", "middle", or "inner".')
```
